```python
# Import dependencies
import pandas as pd
```

```python
# Load in stocks data
stocks = pd.read_csv('stock_data.csv')
stocks.head()
```

```python
cases = pd.read_csv('cases.csv')
cases.head()
```

```python
df = pd.merge(stocks, cases,  how='left', left_on=['country','date'], right_on = ['countryCode','date'])
df = df[['country','date','price','confirmedCases']]
df = df.rename(columns = {
    'confirmedCases' : 'cases'
})

df['id'] = df.index

# Organize columns to match database table
df = df[['id','country','date','price','cases']]

df = df.dropna()

df
```

```python
df.to_csv('stocks_and_cases.csv', index=False, header=False)
```
